### Data Cleaning

```python
# Read all three datasets
import pandas as pd

reactions = pd.read_csv("Reactions.csv")
content = pd.read_csv("Content.csv")
reactiontype = pd.read_csv("ReactionTypes.csv")

```

```python
reactions.head()
```

```python
content.head()
```

```python
reactiontype.head()
```

```python
# drop unused columns
reactions = reactions.drop(["Unnamed: 0" , "User ID"] , axis=1)
content = content.drop(["Unnamed: 0" , "User ID"] , axis=1)
reactiontype = reactiontype.drop("Unnamed: 0" , axis=1)


```

```python
# check data types

print(reactions.info())
print(content.info())
print(reactiontype.info())
```

```python
# find null values


print("-----------" , reactions.isnull().sum())
print("----------", content.isnull().sum())
print("------------", reactiontype.isnull().sum())
```

```python
# Clean URL column 
content = content.dropna()
content.isnull().sum()
```

```python
content.head()
```

```python
reactions.head()
```

```python
# drop nan values from reactions
reactions.isnull().sum()

reactions = reactions.dropna()
```

```python
reactions.head()
```

```python
reactiontype.isnull().sum()
```

```python
# check data types

print(f"reactions:\n \n {reactions.info()}")
print(f"content:\n {content.info()}")
print(f"reactiontype:\n {reactiontype.info()}")
```
